```python
import nltk
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```python
# Tokenization -- paragraphs - sentences- words 

# nltk.download('punkt')
sentences=nltk.sent_tokenize(paragraph)
sentences
```

```python
# example of lemmatization
from nltk.stem import WordNetLemmatizer
# nltk.download('wordnet')
# nltk.download('stopwords')
lemmatizer=WordNetLemmatizer()
lemmatizer.lemmatize('goes')
```

```python
len(sentences)
```

```python
# clean the text, spcl char
import re

corpus=[]
for i in range(len(sentences)):
    review=re.sub('[^a-zA-Z]',' ',sentences[i])
    review=review.lower()
    corpus.append(review)
```

```python
corpus
```

```python
stemmer=PorterStemmer()
```

```python
# Applying stemming
for i in corpus:
    # print(i)
    words=nltk.word_tokenize(i)
    for word in words:
        if word not in set(stopwords.words('english')):
            print(stemmer.stem(word)) 
```

```python
stopwords.words('english')
```

```python
# lemmatization
for i in corpus:
    # print(i)
    words=nltk.word_tokenize(i)
    for word in words:
        if word not in set(stopwords.words('english')):
            print(lemmatizer.lemmatize(word)) 
```

```python
# Bag of words example
from sklearn.feature_extraction.text import CountVectorizer

# Sample documents
documents = [
    "This is the first document.",
    "This document is the second document.",
    "And this is the third one.",
    "Is this the first document?",
]

# Initialize CountVectorizer
vectorizer = CountVectorizer()

# Fit the vectorizer to the documents and transform the documents into a matrix of token counts
X = vectorizer.fit_transform(documents)

# Convert the matrix to an array for better visualization
X_array = X.toarray()

# Get the feature names (unique words)
feature_names = vectorizer.get_feature_names_out()

# Print the matrix and feature names
print("Matrix of token counts:")
print(X_array)
print("\nFeature names:")
print(feature_names)

```

```python
# apply stopwords -- sparse matrix becomes smaller

corpus=[]
for i in range(len(sentences)):
    review=re.sub('[^a-zA-Z]',' ',sentences[i])
    review=review.lower()
    review=review.split()
    review=[lemmatizer.lemmatize(word) for word in review if not word in set(stopwords.words('english'))]
    review=' '.join(review)
    corpus.append(review)
```

```python
cv=CountVectorizer(binary=False)
X=cv.fit_transform(corpus)
cv.vocabulary_
```
